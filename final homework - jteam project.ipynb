```python
service = df[df['업종대분류']=='생활서비스']
service['이용월'] = pd.to_datetime(service['이용월'], format = "%Y%m")
service_Month = service['이용월']
service_Pay = service['카드결제건수(천건)']

plt.plot(service_Month, service_Pay, label = "service", c = 'r')
plt.xlabel("2019.01~2021.08")
plt.ylabel("Pay")
plt.legend()
```

```python
x = service_Month
y = service['카드결제건수(천건)']
plt.xlabel("2019.01~2021.08")
plt.ylabel("Pay")
plt.plot(x,y,'*')
```
