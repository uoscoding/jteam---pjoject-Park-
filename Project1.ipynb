```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel('Card_Consumption.xlsx')

df.head()

FNB = df[df['업종대분류']=='FNB']
ApllianceFurniture = df[df['업종대분류']=='가전/가구']
LeisureCulture = df[df['업종대분류']=='문화여가']

#FNB = FNB.set_index('이용월')



```

```python
FNB.loc[:,"이용월"] = pd.to_datetime(FNB.loc[:,"이용월"], format = "%Y%m")
```

```python
FNB = df[df['업종대분류']=='FNB']
FNB_2019 = FNB[FNB['이용월'] <= 201912]
FNB_2019['이용월'] = pd.to_datetime(FNB_2019['이용월'], format = "%Y%m")
FNB2019_Month = FNB_2019['이용월']
FNB2019_Pay = FNB_2019['카드결제건수(천건)']
```

```python
plt.plot(FNB2019_Month, FNB2019_Pay, label = "FNB", c = 'r')
plt.xlabel("2019.01~2019.12")
plt.ylabel("Pay")
plt.legend()
```

```python
ApllianceFurniture = df[df['업종대분류']=='가전/가구']
AF_2019 = ApllianceFurniture[ApllianceFurniture['이용월'] <= 201912]
AF_2019['이용월'] = pd.to_datetime(AF_2019['이용월'], format = "%Y%m")
AF2019_Month = AF_2019['이용월']
AF2019_Pay = AF_2019['카드결제건수(천건)']
```

```python
plt.plot(AF2019_Month, AF2019_Pay, label = "HomeAplliance/Furniture", c = 'g')
plt.xlabel("2019.01~2019.12")
plt.ylabel("Pay")
plt.legend()
```

```python
LeisureCulture = df[df['업종대분류']=='문화여가']
LC_2019 = LeisureCulture[LeisureCulture['이용월'] <= 201912]
LC_2019['이용월'] = pd.to_datetime(LC_2019['이용월'], format = "%Y%m")
LC2019_Month = LC_2019['이용월']
LC2019_Pay = LC_2019['카드결제건수(천건)']
```

```python
plt.plot(LC2019_Month, LC2019_Pay, label = "Leisure Culture", c = 'b')
plt.xlabel("2019.01~2019.12")
plt.ylabel("Pay")
plt.legend()
```
